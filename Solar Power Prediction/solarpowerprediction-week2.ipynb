```python
# Plot distribution of wind speed
plt.figure(figsize=(10, 6))
sns.histplot(df['wind_speed_900_mb'], bins=50, kde=True)
plt.title('Distrubition of wind Power ')
plt.xlabel('wind speed')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bivariate analysis
# Scatter plot with target features 
plt.figure(figsize=(15, 30))
for i, columns in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    plt.scatter(df[columns], df['generated_power_kw'])
    plt.title(f'{columns} vs Generated Power (KW)')
    plt.xlabel(columns)
    plt.ylabel('Generated Power (KW)')
plt.tight_layout()
plt.show()
```

```python
# Outlier 
plt.figure(figsize=(15, 30))
for i, columns in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    sns.boxplot(df[columns])
plt.show()
```
